4. **Problem** -  There are lines in common between M57 and Vega but they are in absorption in Vega and in Emission in M57. 
**(a)** Identify the lines in common.  *Note that there may be small wavelength shits between the two spectrum because of imperfect wavelength calibration.*

**(b)** Explain why they are in emission in one spectrum and in absorption in the other.  Use your knowledge of Kirkhoffs laws.

*Note - The line at 6880$\mathring A$ is from the Earth's atmosphere and should be ignored*

**This is a result of Kirkhoffs laws. Vega is a hot dense object, a star. At it's core, it is very hot, and is surrounded by relatively much cooler gas. The light is generated at the core and must work it's way through this cold gas before escaping the star and making it's way to us. That gas has a chemical composition where each atom within it has certain quantized values of energy that result in an electron being kicked up an energy level in that atom. The light from the star's core is at all wavelengths, but when it passes through the cooler gas, the cooler gas' electrons will absorb the energy of light of a certain wavelength if it corresponds to the energy needed to kick it up an energy level in that specific atom. This results in absorption lines, while if the light isn't corresponding to one of those quantized values, it passes through to us unaffected. In the case of the planetary nebula, most of the spatial area of the nebula is the cold gas, while the star is just a tiny speck, so most of what is observed is from the cold gas. This is an emission spectrum because the cloud is being 'warmed' by the light from the star and absorbing the light from the same process as above, exciting the electrons, but then they drop down to lower levels, emitting a photon and showing a line in the spectrum there.**

Do your best given your knowledge of how light propogates through media to describe why Jupiter's spectrum is significantly redder than that of the sun.  

*As an aside, note that the very wide absorption lines longward of 6000$\mathring A$ in Jupiter's spectrum are absorption bands from molecules in Jupiter's atmosphere.  These bands absorb light over a wide wavelength range.*

**Jupiter is a gas giant, so the light from the sun gets partially absorbed and re-emitted by Jupiter, whose atmosphere affects how the light comes back to us. Blue light is always more scattered by material in an atmosphere such as dust grains, because it's wavelength is smaller than other visible colors of light, so an object affects it more, because the object is closer to the size of the wave itself. Red light has a wavelength so large that it can pass by such objects mostly unaffected because the objects are so small compared to it's wave size. Jupiter has an atmosphere of dust-grain sized objects, and these scatter light. So the light coming from the sun is as it is, but once it hits Jupiter, the blue light is scattered in random directions, while the red light is mostly unaffected. Therefore, to us, Jupiter seems much redder than the sun, because along the way of the sunlight interacting with Jupiter, the blue light got directed away from us and into random directions that mostly are not toward us at all.**
